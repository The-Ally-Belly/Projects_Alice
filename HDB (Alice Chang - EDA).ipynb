```python
# Resale values by flat types - use groupby and the average resale prices to compute the data

flat_price = hdb_2010.groupby("flat_type")["resale_price"].agg(np.mean).round(decimals=0).reset_index()

# Sort the flat types and prices from highest to lowest

flat_price.sort_values('resale_price', ascending=False, inplace=True)
print("The avg resale prices for each flat type for the period starting from 2010 : ")
flat_price
```

```python
# Convert "month" to datetime format and assign to a new column "year_trans"

from datetime import datetime
list1 = []
list2 = hdb['month'].astype(str).values.tolist()
for i in list2:
    obj = datetime.strptime(i, "%Y-%m")
    list1.append(obj)
hdb['year_trans'] = list1
```

```python
# Convert "lease_commence_date" to datetime format.

list1 = []
list2 = hdb["lease_commence_date"].astype(str).values.tolist()
for i in list2:
    obj = datetime.strptime(i, "%Y")
    list1.append(obj)
hdb["lease_commence_date"] = list1
```

```python
# Create a new column "flat_age" to compute the age of the flat.

hdb["flat_age"] = hdb["year_trans"] - hdb["lease_commence_date"]

# Convert "month" to string, then to integer while removing the 'mm' variable.

hdb['month'] = [int(i[:-3]) for i in hdb['month']]
```

```python
# Check the revised dataframe

hdb.head()
```

```python
# Convert "flat_age" into integer

hdb["flat_age"] = ((pd.to_numeric(hdb["flat_age"].dt.days, downcast="integer"))/365).round(decimals=0)

# Compute remaining_lease

hdb["remaining_lease"] = 99-hdb["flat_age"]
```

```python
# Check the dataframe

hdb.head()
```

```python
# Remove columns 

hdb.drop(columns=['block', 'street_name', 'flat_model'], inplace=True)
```

```python
# Rename 'month' to 'year'

hdb.rename(columns={'month': 'year'}, inplace=True)
```

```python
# Check to see the changes in the columns

hdb.head()
```

```python
# create a new column for 'price_per_sqm'

hdb['price_sqm'] = hdb['resale_price']/hdb['floor_area_sqm']
hdb.head()
```

```python
# use dummy variables to convert 'town' into 'town1' 

from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(hdb.town)
town1 = le.transform(hdb.town)
```

```python
# Insert 'town1' as a new column in the dataframe

hdb['town1'] = town1
```

```python
# Filter the dataframe based on year 2010 onwards

hdb_2010_1 = hdb.loc[hdb.year >= 2010]
hdb_2010_1
```

```python
# Find the average resale price of each town based on 4-room flats from the period 2010 onwards.

hdb_4rm_2010 = hdb_2010_1.loc[hdb_2010_1['flat_type']=='4 ROOM']
top_10_hdb_4rm_2010 = hdb_4rm_2010.groupby('town')['resale_price'].mean().reset_index().sort_values('resale_price',ascending=False).head(10)
print("Average resale prices of top 10 locations for 4-rm HDB since 2010: ")
top_10_hdb_4rm_2010
```

```python
# define top 10 locations for 4-room flats based on resale price

central_area_2010 = hdb_4rm_2010.loc[hdb_4rm_2010['town']=='CENTRAL AREA']
queenstown_2010 = hdb_4rm_2010.loc[hdb_4rm_2010['town']=='QUEENSTOWN']
bt_merah_2010 = hdb_4rm_2010.loc[hdb_4rm_2010['town']=='BUKIT MERAH']
bt_timah_2010 = hdb_4rm_2010.loc[hdb_4rm_2010['town']=='BUKIT TIMAH']
toa_payoh_2010 = hdb_4rm_2010.loc[hdb_4rm_2010['town']=='TOA PAYOH']
kallang_whampoa_2010 = hdb_4rm_2010.loc[hdb_4rm_2010['town']=='KALLANG/WHAMPOA']
clementi_2010 = hdb_4rm_2010.loc[hdb_4rm_2010['town']=='CLEMENTI']
bishan_2010 = hdb_4rm_2010.loc[hdb_4rm_2010['town']=='BISHAN']
marine_parade_2010 = hdb_4rm_2010.loc[hdb_4rm_2010['town']=='MARINE PARADE']
geylang_2010 = hdb_4rm_2010.loc[hdb_4rm_2010['town']=='GEYLANG']
```

```python
# Stack the datasets for the top 10 locations for 4-room hdb flats

top_hdb_4rm = pd.concat([central_area_2010, queenstown_2010, bt_merah_2010, bt_timah_2010, toa_payoh_2010, 
                         kallang_whampoa_2010,clementi_2010, bishan_2010, marine_parade_2010, geylang_2010])
top_hdb_4rm
```

```python
# plot line graph between resale price and location for 4-room flats, the most popular HDB flat type.

fig=plt.gcf()
fig.set_size_inches(11,8);
sns.lineplot(data=top_hdb_4rm, x='year', y='resale_price', hue='town', estimator='mean', ci=None, palette='bright');

plt.title('Average Resale Prices for 4-Room Flat by Top 10 Locations from 2010 onwards');
plt.xlabel('year');
plt.ylabel('Resale Price');
```

```python
# The flat age and price per sqm for 2010 onwards
# hdb_2010_1

fig=plt.gcf()
fig.set_size_inches(11,8);
sns.lineplot(data=hdb_2010_1, x='flat_age', y='price_sqm');
plt.title('How age of HDB flats affects price per sqm');
```

```python
# Find the average resale prices of each town based on 4-room flats from the period 2019 onwards.

hdb_4rm_2019 = hdb_2010_1.loc[(hdb_2010_1['flat_type']=='4 ROOM') & (hdb_2010_1['year']>= 2019)]
top_10_hdb_4rm_2019 = hdb_4rm_2019.groupby('town')['resale_price'].mean().reset_index().sort_values('resale_price',ascending=False).head(10)
top_10_hdb_4rm_2019
```

```python
# Identify 3 preferred locations(towns)

hdb_4rm_Bt_Timah = hdb_4rm_2019.loc[hdb_4rm_2019['town']=='BUKIT TIMAH']
hdb_4rm_Toa_Payoh = hdb_4rm_2019.loc[hdb_4rm_2019['town']=='TOA PAYOH']
hdb_4rm_Bishan = hdb_4rm_2019.loc[hdb_4rm_2019['town']=='BISHAN']
```

```python
# Filter the list for Bukit Timah by flat age (<=25 years) and price (<=$600K)

criteria = (hdb_4rm_Bt_Timah['flat_age'] <=25.0) & (hdb_4rm_Bt_Timah['resale_price'] <= 600000.0)
Bt_Timah_25yrs = hdb_4rm_Bt_Timah.loc[criteria]
Bt_Timah_25yrs
```

```python
# Filter the list for Toa Payoh by flat age (<=25 years) and price (<=$600K)

criteria = (hdb_4rm_Toa_Payoh['flat_age'] <=25.0) & (hdb_4rm_Toa_Payoh['resale_price'] <= 600000.0)
Toa_Payoh_25yrs = hdb_4rm_Toa_Payoh.loc[criteria]
Toa_Payoh_25yrs
```

```python
# Narrow the list for Toa Payoh to minimum of 100 sqm for floor area. 

Toa_Payoh_25yrs_100sqm = Toa_Payoh_25yrs.loc[Toa_Payoh_25yrs['floor_area_sqm'] >= 100.0]
Toa_Payoh_25yrs_100sqm
```

```python
# Seek out the block numbers and street name for Toa Payoh's selected list by index.

Toa_Payoh_List = df5[df5.index.isin(Toa_Payoh_25yrs_100sqm.index)]
Toa_Payoh_List
```

```python
# Zoom in on the block number for the best fit

print("Blocks recommended for Toa Payoh 4rm flat (Model A): ")
Toa_Payoh_List['block'].value_counts()
```

```python
# Zoom in on relevant street names for Toa Payoh to find the best fit

print("Street names recommended for Toa Payoh 4rm flat (Model A): ")
Toa_Payoh_List['street_name'].value_counts()
```
